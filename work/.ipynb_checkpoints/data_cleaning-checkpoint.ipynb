```python
import numpy as np;
import pandas as pd;
import matplotlib.pyplot as plt;
import seaborn as sns;
```

```python
df = pd.read_csv("zomato.csv")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.drop("dish_liked",axis=1,inplace=True)
```

```python
# Duplicated Values
print(f"duplicated values : {df.duplicated().sum()}")
df.drop_duplicates(inplace=True)
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df.head()
```

```python
df.rename(columns={"approx_cost(for two people)":"cost","listed_in(type)":"type","listed_in(city)":"city"},inplace=True)
```

```python
# clean rate column
rateColumn = df["rate"]

rateColumn = rateColumn.apply(lambda x : x.split("/")[0])
rateColumn = rateColumn.apply(lambda x : 0.0 if x == "NEW" or x =="-" else x)

rateColumn.unique()
```

```python
df["rate"] = rateColumn
df["rate"].head()
```

```python
df.info()
```

```python
# Adjust datatype of columns to correct datatype
df = df.astype({"rate":"float64"})
```

```python
cost = df["cost"]
cost = cost.astype(str)
cost = cost.apply(lambda X : X.replace(",",""))
cost = cost.astype("int64")
cost.unique()
df["cost)"] = cost
```

```python
#Copy Dataframe to another frame
df_copy = df.copy()
```

```python
# we see listed_in(city)) and location are same so we drop one column and
df_copy.drop(["city","address","url","menu_item","phone"],axis=1,inplace=True)
```

# Geospatial Analysis

```python
from geopy.geocoders import Nominatim
```

```python
geolocator=Nominatim(user_agent="app")
```

```python
locations = pd.DataFrame({"Name":df["location"].unique()})

lat = []
lon = []
for location in locations["Name"]:
    location = geolocator.geocode(location)

    if location == None:
        lat.append(np.nan)
        lon.append(np.nan)
    else:
        lat.append(location.latitude)
        lon.append(location.longitude)

locations["lat"] = lat
locations["lon"] = lon
```

```python
locations.head()
```

```python
rest_locations = pd.DataFrame(df["location"].value_counts().reset_index())
```

```python
rest_locations.columns = ["Name","count"]
rest_locations
```

```python
print(locations.shape)
print(rest_locations.shape)
```

```python
restaurant_locations = rest_locations.merge(locations, on="Name", how = "left").dropna()
restaurant_locations.head()
```

```python
restaurant_locations["count"].max()
```
